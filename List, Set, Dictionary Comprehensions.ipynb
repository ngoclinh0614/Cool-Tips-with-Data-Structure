## List comprehension

```python
#Instead of creating a list in this old way:
result = []
for i in range(10): #for val in collection
    if i%2 == 0:    #if condition
        result.append(i) #list.append(expr)
result
```

```python
#Can use a simple list like this
result = [i for i in range(10) if i%2==0] #list = [expr for val in collection if condition]
result
```

```python
#Convert into uppercase
list_a = ['Nguyen', 'Tran', 'Le', 'Doan']
list_b = [i.upper() for i in list_a if len(i)>2]
list_b
```

## Set & Dict comprehension

```python
#Similar to list comprehension:
#dict_comp = {key-expr : value-expr for value in collection if condition}
dict_a = {value:key for value,key in enumerate(list_a)}
dict_a 
```

```python
#set_comp = {expr for value in collection if condition}
set_a = {i for i in range(10) if i%2==1}
set_a
```

```python
a = ['Dao', 'Kieu', 'Van', 'Nhi'] #using map to define len of every element in list/set
list(map(len, a))
set(map(len,a))
```

## Nested list comprehension

```python
name = [['Nguyenn', 'Trann', 'Hoanng', 'Le'], ['Gianng', 'Dao', 'Minnh', 'Uyenn']]
#To filter name that have more than 1 'n' in it, we often do:
filter_name =[]
for list_name in name:
    for names in list_name:
        if names.count('n') >1:
            filter_name.append(names)
filter_name
```

```python
#Instead, can using nested list comprehension:
filter_name =[names for list_name in name for names in list_name if names.count('n')>1]
filter_name

### NOTE: the order of 'for' in nested list must be same as when we wrote long n specific code
```

```python
#Use nested list comprehension to flatten tuple in list:
a =[(1,2,3), (4,5,6), (7,8,9)]
flatten_a = [i for tup in a for i in tup]
flatten_a
```

```python
#Or product a list of lists
list_of_list = [[i for i in tup] for tup in a]
list_of_list
```
