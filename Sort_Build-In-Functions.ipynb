# Sort

```python
#Normal sort
a = [4,5,9,23,56,78,8]
a.sort()
a
```

```python
#Sort by key, eg: key = length of list
b = ['Nguyen', "Tran", 'Hoang', 'Le', "Do", 'Mac']
b.sort(key = len, reverse= True)
b
```

```python
#Cool tips to reverse: using slicing with ::-1
a[::-1]
```

### Zip

```python
seq1 = ['Nguyen', 'Tran', 'Hoang']
seq2 = ['Thao', "Van", 'Nhu']
zipped = zip(seq1, seq2)
print(list(zipped))
```

```python
for a, (i,j) in enumerate(zip(seq1, seq2)):
    print(a+1, i, 'is last name of', j)
```

```python
# using zip to UNZIP
list_b = [('Nguyen', 'Thao'), ('Tran', 'Van'), ('Hoang', 'Nhu')]
last_name, first_name = zip(*list_b) #using * before list to unzip it!!!
```

```python
last_name
```

```python
first_name
```

# Dictionary

```python
name = {1: 'Nguyen', 2: 'Tran', 3:'Le', 4:'Do', 5:'Hoang'}
name
```

```python
#add to dict
name[6] ='Van'
name
```

```python
#Delete using DEL 
del name[1]
name
```

```python
name_pop = name.pop(4)
name, name_pop
```

```python
#Insert a dict to a dict using UPDATE
name.update({8:'Vu', 9: 'Dao'})
name
```

```python
#create a dict from sequence:
dict_test ={}
for key, value in zip(range(5), reversed(range(5))):
    dict_test[key] = value
dict_test
```

```python
#or just using dict()
dict_test_2 = dict(zip(range(10), reversed(range(10))))
dict_test_2
```
